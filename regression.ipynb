```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataPair = 20 # Number of points
#Using uniform distribution between 0 and 1 for X
x=np.random.uniform(0,1,dataPair)
x
```

```python
#Generating Gaussian Noise
N=np.random.normal(0,1,size=dataPair)
N
fig = plt.figure()
plt.plot(x,N, 'b-')
```

```python
#Generating Sine Wave
y = np.sin(2*np.pi*x)+ 0.1*N
y
```

```python
#plt.plot(x,y, 'b-')
plt.scatter(x,y)
```

```python
y
newarr = np.array_split(y, 2)
ytrain=newarr[0]
np.sort(ytrain)
ytest=newarr[1]

ytest.sort()
print("Training set of y -values", ytrain)
print('Testing set of y-values ',ytest)
```

```python
x
arr = np.array_split(x, 2)
xtrain=arr[0]
xtrain.sort()
xtest=arr[1]
xtest.sort()
print("Training set of x-values", xtrain)
print('Testing set of x-values ',xtest)
```

```python

#Drawing test vs train 
plt.figure
plt.plot(xtrain, ytrain, 'bo',xtest, ytest, 'ro')
plt.legend(('Training data', 'Testing data'))
```

```python

def coefficients(M) :
    coeff = np.polyfit(xtrain, ytrain, M) # Fitting the polynomial and finding coefficients
    print('Coefficients for M=',M,'are', coeff)
    p = np.poly1d(coeff)# Polynomial construction
    y_pred = p(xtrain)
    return y_pred
coefficients(0)
coefficients(1)
coefficients(3)
coefficients(9)
```

```python
#Computing the Mean Square Error
y_pred=coefficients(0)
err = 0.5*(np.sum((y_pred-ytrain)**2))
print ('Error is ',err) 
```
